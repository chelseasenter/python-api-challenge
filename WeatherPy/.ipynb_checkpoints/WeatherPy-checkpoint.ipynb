```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint


# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=2000)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=2000)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
```

```python
# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# perform a weather check on each city using a series of successive API calls.
# Include a print log of each city as it'sbeing processed (with the city number and city name).

city_data_dict = {}
search = 0

for sets in range(1,13):
    successful = 0
    while successful < 50:

        url = f'http://api.openweathermap.org/data/2.5/weather?q={cities[search]}&appid={weather_api_key}&units=imperial'
        response = requests.get(url).json()
        if response['cod'] == '404':
            print(f'{cities[search]} not found, skipping...')
        else:
            print(f"Processing Record {successful+1} of Set {sets} | {cities[search]}")
            city_info = []
            city_info.append(response['name'])
            city_info.append(response['coord']['lat'])
            city_info.append(response['coord']['lon'])
            city_info.append(response['main']['temp_max'])
            city_info.append(response['main']['humidity'])
            city_info.append(response['clouds']['all'])
            city_info.append(response['wind']['speed'])
            city_info.append(response['sys']['country'])
            city_info.append(response['dt'])
            
            city_data_dict[search] = city_info
            
            successful = successful + 1
        search = search + 1
        time.sleep(1)
```

```python
url = f'http://api.openweathermap.org/data/2.5/weather?q=chapleau&appid={weather_api_key}&units=imperial'
response = requests.get(url).json()
response
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# print the dictionary of city information
# pprint(city_data_dict)
# print(city_data_dict)

#create the dataframe
data_df = pd.DataFrame.from_dict(city_data_dict, orient='index', 
                       columns=['Name', 
                                'Lat', 
                                'Lng', 
                                'Max Temp', 
                                'Humidity', 
                                'Cloudiness', 
                                'Wind Speed', 
                                'Country', 
                                'Date']).reset_index(drop=True)

data_df
```

```python
# Output File (CSV)
output_data_file = "../output_data/cities.csv"
data_df.to_csv(output_data_file, index = False, header=True)
```

## Linear Regression

```python
# setup the northern and southern dataframes
northern_df = data_df.loc[data_df['Lat'] > 0]
southern_df = data_df.loc[data_df['Lat'] < 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
n_x_lat = northern_df['Lat']
n_y_temp = northern_df['Max Temp']


plt.scatter(n_x_lat, n_y_temp, marker='o', facecolors='red', alpha=0.50)
plt.title('City Latitude Vs. Maximum Temperature')
plt.xlabel('Latitude')
plt.ylabel('Maximum Temperature (F)')

(slope, intercept, rvalue, pvalue, stderr) = linregress(n_x_lat, n_y_temp)
regress_values = n_x_lat * slope + intercept
plt.plot(n_x_lat,regress_values,"maroon")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(0,-40),fontsize=12, color="maroon")

plt.savefig("../output_data/northHumid.png")
```
